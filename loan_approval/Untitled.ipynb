```python
df_train['loan_grade'] = label.fit_transform(df_train['loan_grade'])
```

```python
df_test['loan_grade'] = label.transform(df_test['loan_grade'])
```

```python
df_test['cb_person_default_on_file'] = label.transform(df_test['cb_person_default_on_file'])
```

```python
X = df_train.drop('loan_status', axis = 1)
y = df_train['loan_status']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
df_test = scaler.transform(df_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
import xgboost as xgb
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
model1 = xgb.XGBClassifier(
    n_estimators=650,           
    max_depth=6,                
    learning_rate=0.08,         
    subsample=0.8,              
    colsample_bytree=0.8,       
    gamma=1,                    
    reg_alpha=0.2,             
    reg_lambda=1,               
    use_label_encoder=False,    
    eval_metric='logloss',      
    random_state=42             
)
```

```python
model1.fit(X, y)
```

```python
y_pred = model1.predict_proba(df_test)
```

```python
submission = pd.read_csv('sample_submission.csv')
```
